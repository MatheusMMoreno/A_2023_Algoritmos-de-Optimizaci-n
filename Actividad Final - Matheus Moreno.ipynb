```python
df = pd.read_excel("Datos problema doblaje(30 tomas, 10 actores).xlsx", index_col ='Toma')

#El objetivo es buscar  minimizar el numero de actores por dia tomando en cuenta que se puede hacer 6 tomas por dia

def generate_solution(df,splits =6 ):# particiones con restricion de tomas por dia

    tomas = list(df.index)

    dfs = []    

    indexes = [ tomas[i:splits+i] for i in range(0,len(tomas),splits)]


    for i in indexes:
        df_stage = df.loc[i,:]
        
        dfs.append(df_stage)

    return dfs ,indexes

def pago(dfs ): # calculo de cuantos actores fueron pagos para las 30 tomas
    pago_actores = 0

    for i in range(len(dfs)):
        pago_dia = len(dfs[i].loc[:,dfs[i].sum(axis=0)>0].columns)
        
        pago_actores += pago_dia


    return pago_actores

dfs, indexes = generate_solution(df, splits=6 )


print('random solution cost:',pago(dfs))

changes= []

for n in range(len(dfs)):
    # iterar los dfs

    for i in range(len(dfs[n])):
        # iterar por las lineas del  df referencia 
        
        for j in range(len(dfs)):
            # para cada linea del df de referencia iterar por las lineas de los otros dfs     

            if j == n: # no tomar en cuenta mismo df
                continue

            for z in range(len(dfs[j])):
                # iterar las lineas del df de comparacion
                dfs_copy = [df_stage.copy() for df_stage in dfs]
                
                # armacenar indices
                index_base = dfs_copy[n].index[i]
                index_test = dfs_copy[j].index[z]
                
                row1 = dfs_copy[n].loc[index_base].copy()
                row2 = dfs_copy[j].loc[index_test].copy()
                
                # Swap 
                dfs_copy[n].loc[index_base] = row2
                dfs_copy[j].loc[index_test] = row1

                dfs_copy[n].rename(index = {index_base:index_test},inplace=True)
                dfs_copy[j].rename(index = {index_test:index_base},inplace=True)
                

                #calcular nuevo custo
                pago_total = pago(dfs_copy)
                
                #calcular  custo
                pago_actual = pago(dfs)
                
        
                if pago_total < pago_actual:
                    changes.append({
                        'action': 'swap',
                        'old paid' :pago_actual,
                        'new payment' :pago_total,
                        'df1':n,
                        'index df1':index_base,
                        'index df2':index_test,
                        'df2':j
                    })
                    dfs[n].loc[index_base] = row2
                    dfs[j].loc[index_test] = row1

                    dfs[n].rename(index = {index_base:index_test},inplace=True)
                    dfs[j].rename(index = {index_test:index_base},inplace=True)
                else:
                    continue

print('final solution cost:',pago(dfs))

for df_display in dfs:
    display(df_display)
```

```python
df = pd.read_excel("Datos problema doblaje(30 tomas, 10 actores).xlsx", index_col ='Toma')

def generate_solution(df,splits =6 ):# particiones con restricion de tomas por dia

    tomas = list(df.index)

    dfs = []    

    indexes = [ tomas[i:splits+i] for i in range(0,len(tomas),splits)]


    for i in indexes:
        df_stage = df.loc[i,:]
        
        dfs.append(df_stage)

    return dfs ,indexes

dfs, indexes = generate_solution(df, splits=6 ) #create a solution

def generate_vecino_solution(dfs):
    #hace una copia de los dfs original
    dfs_copy = [df_stage.copy() for df_stage in dfs]

    #genera dos grupos que son diferentes
    group1 = random.randint(0, len(dfs) - 1)
    group2 = random.randint(0, len(dfs) - 1)
    while group1 == group2:
        group2 = random.randint(0, len(dfs) - 1)

    #elije indices aleatorios  de group1 y group2
    index1 = dfs[group1].index[random.randint(0, len(dfs[group1]) - 1)]
    index2 = dfs[group2].index[random.randint(0, len(dfs[group2]) - 1)]

    #se copian las lineas y haz el swap
    row1 = dfs_copy[group1].loc[index1].copy()
    row2 = dfs_copy[group2].loc[index2].copy()


    dfs_copy[group1].loc[index1] = row2
    dfs_copy[group2].loc[index2] = row1

    dfs_copy[group1].rename(index = {index1:index2},inplace=True)
    dfs_copy[group2].rename(index = {index2:index1},inplace=True)

    return dfs_copy

def pago(dfs):
    pago_actores = 0

    for df in dfs:
        pago_dia = len(df.loc[:, df.sum(axis=0) > 0].columns)
        pago_actores += pago_dia

    return pago_actores

def local_search(df, splits=6, max_iterations=1000):
    #genera una solucion
    dfs, _ = generate_solution(df, splits=splits)
    best_cost = pago(dfs)
    best_solution = dfs.copy()

    #hace 1000 pruebas haver si la solucion mejora
    for _ in range(max_iterations):
        new_solution = generate_vecino_solution(best_solution)
        new_cost = pago(new_solution)

        if new_cost < best_cost:
            best_solution = new_solution.copy()
            best_cost = new_cost

    return best_solution, best_cost

dfs, cost = local_search(df, splits=6)
print("Final solution cost:", cost)

for df_display in dfs:
    display(df_display)
```
