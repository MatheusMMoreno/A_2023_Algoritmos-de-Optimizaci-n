```python
import math
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
f = lambda X : X[0]**2 + X[1]**2
df = lambda X : [2*X[0], 2*X[1]]
```

```python
#mapa de niveles

resolucion = 100
rango = 2.5

X=np.linspace(-rango, rango, resolucion)
Y=np.linspace(-rango,rango,resolucion)
Z=np.zeros((resolucion,resolucion))

for ix,x in enumerate(X):
    for iy,y in enumerate(Y):
        Z[iy,ix]=f([x,y])

contour = plt.contourf(X,Y,Z,resolucion)
cbar = plt.colorbar()

P=[random.uniform(-rango,rango),random.uniform(-rango,rango)]
plt.plot(P[0],P[1],"o",c="red")

TA = 0.05

for _ in range(500):
    grad = df(P)
    P[0], P[1] = P[0] - TA *grad[0], P[1] - TA * grad[1]
    plt.plot(P[0],P[1],"o",c="green")

print("Solucion:", P, f(P))
```

```python
#La funcion
f = lambda X :np.sin(1/2  *X[0]**2 - 1/4 * X[1]**2 +3) * np.cos(2 * X[0]+1 -np.e**X[1])
```

```python
def df(PUNTO):
    h = 0.01
    T = np.copy(PUNTO)
    grad = np.zeros(2)

    for it, th in enumerate(PUNTO):
        T[it] = T[it] + h # hace un pequeno incremento en el iten del array mientras mantiene el otro iten sin cambio
        grad[it] = (f(T) - f(PUNTO))/h # hace el calculo de la tangente cuando h esta cerca de cero

    return grad
```

```python
#mapa de niveles

plt.figure(figsize=(8,8))

resolucion = 100
rango = 2.5

X=np.linspace(-rango, rango, resolucion)
Y=np.linspace(-rango,rango,resolucion)
Z=np.zeros((resolucion,resolucion))

for ix,x in enumerate(X):
    for iy,y in enumerate(Y):
        Z[iy,ix]=f([x,y])

contour = plt.contourf(X,Y,Z,resolucion)
cbar = plt.colorbar()

P=[random.uniform(-rango,rango),random.uniform(-rango,rango)]
plt.plot(P[0],P[1],"o",c="red")

TA = 0.05

for _ in range(1000):
    grad = df(P)
    P[0], P[1] = P[0] - TA *grad[0], P[1] - TA * grad[1]
    plt.plot(P[0],P[1],"o",c="green")

    if _ % 10 == 0:
        TA *= 0.99 #ajuste en la taza de aprendisaje

plt.plot(P[0],P[1],"o",c="orange")

print("Solucion:", P, f(P))
```
